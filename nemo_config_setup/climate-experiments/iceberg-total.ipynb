### Visualize iceberg trajectories

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.path as mpath
import glob
```

#### Load files

```python
run_dir   = '/gws/nopw/j04/anthrofail/birgal/NEMO_AIS/output/CESM2_tune0_ens1011/' #trajectories0/'
```

```python
flist_icb = np.sort(glob.glob(f'{run_dir}restart/*restart_icb.nc'))
```

```python
ds = xr.open_dataset(flist_icb[100])
# ds = xr.open_dataset(run_dir)
```

```python
ds.virtual_area.where(np.abs(ds.uvel)>=1e-3).plot(figsize=(10,4),ylim=(80,400),vmin=0,vmax=1e8)
```

```python
# ds.stored_ice.isel(c=0).plot(figsize=(10,4),ylim=(80,400), vmin=0, vmax=1e8)
ds.virtual_area.plot(figsize=(10,4),ylim=(180,380),vmin=0,vmax=1e8)
```

```python
icb_total = np.zeros(len(flist_icb))
for f, file in enumerate(flist_icb):
    print(file)
    ds = xr.open_dataset(file)
    icb_total[f] = ds.n.size
```
