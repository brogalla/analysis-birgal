```python
ts_west = xr.open_dataset(f'{base_dir}output/ERA5_test6/timeseries-WestAntarctica.nc')
```

```python
region_names     = ['pine_island','pine_island_bay','abbot','amundsen_sea','bellingshausen_sea', 'dotson_crosson', 'getz', 'thwaites']
timeseries_types = []
var_names = ['bwtemp']
for region in region_names:
   for var in var_names:
      timeseries_types.append(f'{region}_{var}')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,5))

for tstype in timeseries_types:
    ax.plot(ts_west.time_centered, ts_west[tstype], label=tstype.split('_bwtemp')[0])

ax.legend(loc=(1.05, 0.4), frameon=False, fontsize=12)
ax.set_ylabel('Conservative Bottom Temperature (C)')
```
