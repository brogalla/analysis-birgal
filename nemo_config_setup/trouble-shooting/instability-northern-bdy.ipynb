# Instability near northern boundary of domain and tip of south america

Possibly due to:
- 1 rimwidth boundary
- increase in grid size towards the northern part of the domain

```python
import xarray as xr
import cmocean
import matplotlib.pyplot as plt
```

```python
run_folder = '/gws/nopw/j04/anthrofail/birgal/NEMO_AIS/output/ERA5_test4/'
gridT = xr.open_dataset(f'{run_folder}eANT025.L121_1m_19790101_19791231_grid_T.nc')
gridW = xr.open_dataset(f'{run_folder}eANT025.L121_1m_19790101_19791231_grid_W.nc')
```

Vertical velocity current setup:

```python
fig, ax = plt.subplots(1,1, figsize=(20,4))
gridT.thetao.isel(time_counter=6,deptht=10).plot(ax=ax)
ax.set_ylim(430, 453)
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,4))
gridT.mldr10_1.isel(time_counter=6).plot(ax=ax, vmin=0, vmax=500)
ax.set_ylim(430, 453)
```

```python
fig, ax = plt.subplots(1,1, figsize=(20,4))
gridW.wo.isel(time_counter=6, depthw=20).plot(ax=ax, vmin=-2e-4, vmax=2e-4, cmap=cmocean.cm.balance)
ax.set_ylim(430, 453)
```
